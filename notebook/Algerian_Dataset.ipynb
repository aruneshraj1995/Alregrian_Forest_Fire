## Data Cleaning

```python
##missing Values
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

The Dataset is converted into two sets based on Regaion from 122th index. We can make a new colum based on the Region

1: "Bejaia Region Dataset"
2: "Sidi-Bel Abbes Region Dataset"

Add new column with Region

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
dataset=df
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.info()
```

- convert Object type of region from float to int

```python
dataset[["Region"]]=dataset[["Region"]].astype(int)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

- removing missing values

```python
dataset=dataset.dropna().reset_index(drop=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
dataset.iloc[[122]]
```

```python
# remove the 122nd row
dataset=dataset.drop(122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column names

dataset.columns=dataset.columns.str.strip()
dataset.columns
```

```python
dataset.info()
```

## change the required columns as int datatype

```python
dataset.columns
```

```python
dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
dataset.info()
```

```python
dataset[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']]=dataset[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
dataset.info()
```

```python
# save the cleaned dataset
dataset.to_csv('Algerian_forest_fire_fresh_dataset.csv', index=False)
```

## Exploratory Data Analysis

```python
#drop day, month & year

dataset_copy=dataset.drop(['day', 'month', 'year'], axis=1)
```

```python
dataset_copy.head()
```

```python
# categories in classes
dataset_copy['Classes'].value_counts()
```

```python
# encoding of the categories to classes
dataset_copy['Classes']=np.where(dataset_copy['Classes'].str.contains('not fire'),0,1)
```

```python
dataset_copy.tail(25)
```

```python
dataset_copy['Classes'].value_counts()
```

- Density plot for all features

```python
plt.style.use('seaborn')
dataset_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

- Percentage of piechart

```python
percent=dataset_copy['Classes'].value_counts(normalize=True)*100
```

- plotting pie chart

```python
classlabels=['Fire', 'Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percent,labels=classlabels, autopct='%1.1f%%')
plt.title("pie chart of Classes")
plt.show()
```

- correlation

```python
## Monthly Fire Analysis
dftemp=dataset.loc[dataset['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dataset)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Sidi- Bel Regions",weight='bold')
```
